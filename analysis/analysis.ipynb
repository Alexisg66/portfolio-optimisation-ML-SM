## Set up portfolio modelling object

```python
import os

cwd_path = os.path.normpath(os.getcwd()).split(os.sep)

if (cwd_path[-1] == 'analysis') & (cwd_path[-2]=='Diss_modularised'):
    os.chdir('..')

from dotenv import load_dotenv
import os
from analysis import portfolio
from datetime import datetime as dt

load_dotenv('.env')

port = portfolio.Portfolio(os.environ['DB_PATH'])

port.execute(features=['RSI','OBV','MACD','%K','%D'])
```
